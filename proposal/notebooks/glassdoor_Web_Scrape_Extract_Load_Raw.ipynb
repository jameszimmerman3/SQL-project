```python
# COMMENT: Load the results.json data into a DataFrame
import pandas as pd
import json

# Read JSON content
with open("results.json", "r") as f:
    json_data = json.load(f)

# Convert to DataFrame
df_glassdoor = pd.json_normalize(json_data)

# Preview the first few rows
df_glassdoor.head()
```

```python
# COMMENT: Trim down the JSON to relevant Glassdoor metadata fields
columns_to_keep = [
    "metadata.ogTitle",
    "metadata.ogDescription",
    "metadata.description",
    "metadata.language",
    "metadata.sourceURL",
    "metadata.statusCode"
]

df_cleaned_glassdoor = df_glassdoor[columns_to_keep].copy()

# COMMENT: Preview the cleaned metadata
df_cleaned_glassdoor.head()
```

```python
# COMMENT: Set up SQLAlchemy connection to PostgreSQL
from sqlalchemy import create_engine

pg_user = "postgres"
pg_password = "isba_4715"
pg_host = "isba-dev-02.cszeawikopcz.us-east-1.rds.amazonaws.com"
pg_port = "5432"
pg_database = "postgres"

db_url = f"postgresql+psycopg2://{pg_user}:{pg_password}@{pg_host}:{pg_port}/{pg_database}"
engine = create_engine(db_url)
```

```python
# COMMENT: Push Glassdoor metadata to PostgreSQL (public.glassdoor_metadata)
df_cleaned_glassdoor.to_sql(
    "glassdoor_metadata",
    engine,                # reuse your earlier SQLAlchemy engine
    schema="public",
    if_exists="replace",   # overwrite for now
    index=False
)

print("✅ Glassdoor metadata successfully loaded to public.glassdoor_metadata")
```
